```python
df = df.set_index("Unnamed: 0")
df.head()
```

```python
y = df.label 
```

```python
df.drop("label", axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], y, test_size=0.33, random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
count_vectorizer = CountVectorizer(stop_words='english') # ignore the stopwords
count_train = count_vectorizer.fit_transform(X_train)  # Get the training vectors # extract features from text
count_test = count_vectorizer.transform(X_test)# produce model output
```

```python
print('Vocabulary: ')
print(count_vectorizer.vocabulary_)
```

```python
print(count_train.A.shape)# A provide row and column value
print(X_train.shape) # print row
print(len(count_vectorizer.get_feature_names()))# 
```

```python
print(count_vectorizer.get_feature_names())
```

```python
count_df = pd.DataFrame(count_train.A, columns=count_vectorizer.get_feature_names())
```

# Using Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB # multiple variable 
```

```python
clf = MultinomialNB() # Create multinomial naive Bayes object
```

```python
from sklearn import metrics
```

```python
clf.fit(count_train, y_train)
pred = clf.predict(count_test)
pred
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
```
